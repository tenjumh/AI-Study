# Fourier Transform
- 신호처리, 음성, 통신 분야에서 뿐만 아니라 영상처리(Computer Vision)에서도 중요한 개념
- Computer Vision에서는 영상 -> 주파수 성분으로 변환하여 분석 및 처리
- 또한 필터링 연산을 FFT(Fast Fourier Transform)을 이용하여 고속으로 구현

# 1. Fourier Transform - 직관적 이해
- 임의의 입력 신호를 다양한 주파수를 갖는 주기함수(sin, cos)들의 합으로 분해하여 표현
- 사용하는 주기함수는 sin, cos삼각함수이며 Fourier Transform은 고주파수 ~ 저주파수까지 다양한 주파수 대역을 삼각함수들로 분해


![푸리에 변환](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%ED%91%B8%EB%A6%AC%EC%97%90%20%EB%B3%80%ED%99%98.JPG?raw=True)
파란 색 신호들을 합치면 붉은 색 신호가 됨

-입력신호 : 붉은 색 신호<Br>
 *전파,음성 선분은 시간축(t)에 대한 정의된 신호
 *이미지 등은 공간축(x and y)에 대한 정의된 신호

-푸리에 변환을 통해 얻어진 주기함수성분 : 파란 색 신호<Br>
 *통신, 음성에서는 FT를 통해 Time domain -> Frequency domain 변환<Br>
 *영상에서는 FT를 통해 Spatial domain -> Frequency domain 변환

# 2. Fourier Transform - 수식적 이해
- 수식
![푸리에 변환 수식](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%ED%91%B8%EB%A6%AC%EC%97%90%20%EB%B3%80%ED%99%98%20%EC%88%98%EC%8B%9D.JPG?raw=True)
여기서 j-허수단위, f(x)-입력 신호, e-주파수 u인 주기함수 성분, F(u)-주기함수 성분의 계수
- 해석 : 
 - 식(1) Inverse Fourier Transform : 입력신호 f(x)가 e들의 합으로 표현(분해) - e는 f(x)를 구성하는 주기함수 성분
 - 식(2) Fourier Transform : f(x)를 주기함수 성분으로 분해했을 때의 계수 F(u) - 해당 주기함수 성분의 강도(amplitude)를 나타냄
 - e 해석을 위해 오일러공식(Euler's formula)가 필요
 - 오일러 공식은 복소지수함수 -> 삼각함수로 변환하는 식
![오일러공식 수식](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%EC%98%A4%EC%9D%BC%EB%9F%AC%EA%B3%B5%EC%8B%9D%20%EC%88%98%EC%8B%9D.JPG?raw=True)
 이를 이용하여 e는 실수부가 cos(2πux), 허수부 sin(2πux)인 주기함수
![오일러공식 푸리에 수식](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%EC%98%A4%EC%9D%BC%EB%9F%AC%EA%B3%B5%EC%8B%9D%20%ED%91%B8%EB%A6%AC%EC%97%90%20%EC%88%98%EC%8B%9D.JPG?raw=True)

# 3. 영상신호에서의 Fourier Transform
- 상기 1,2는 1차원 신호일 때의 공식이며 이미지는 2차원(x축 방향의 변화, y축 방향읠 변화가 포함된) 신호
- 2차원에서 Fourier Transform은 다음과 같다.
![영상처리 푸리에 변환 수식](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%EC%98%81%EC%83%81%EC%B2%98%EB%A6%AC%20%ED%91%B8%EB%A6%AC%EC%97%90%20%EB%B3%80%ED%99%98%20%EC%88%98%EC%8B%9D.JPG?raw=True)
여기서 F(u,v)는 x축 방향으로 주파수(Frequency) u, y축 방향으로 v인 주기함수 성분 계수
- 이미지는 연속(continuous)이 아닌 이산(discrete) 신호이며 한정된 유한 구간 신호이다.
- W x H 크기의 이미지 f(x,y)에 대한 이산푸리에 변환은 다음과 같다.
![이산 푸리에 변환 수식](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%EC%9D%B4%EC%82%B0%20%ED%91%B8%EB%A6%AC%EC%97%90%20%EB%B3%80%ED%99%98%20%EC%88%98%EC%8B%9D.JPG?raw=True)
식(7)에서 e는 x축 방향으로 주파수가 u/W, y축 방향으로 주파수가 v/H인 주기함수이다.
즉 x축 또는 y축을 시간축으로 놓고 좌표 변화에 따른 이미지 픽섹의 밝기 변화를 신호로 생각하면 된다. <br>
(저주파수: 밝기 변화가 낮은곳, 고주파수: 밝기 변화가 높은 곳 - 이를 통해 저주파수를 버리면 엣지를 찾을 수 있고, 고주파수를 버리면 Smoothing된다.)

# 4. 푸리에 스펙트럼(spectrum)과 페이즈(phase)
-F(u,v)는 복수소이며 실수부와 허수부로 구성<br>
F(u,v) = R(u,v)+jI(u,v)<br>
***이때, 복소수 F(u,v)의 크기 <br>
|F(u,v)|를 푸리에 변환의 spectrum 또는 magnitude라고 함,<br> 
F(u,v)의 각도 Φ를 phase라고 함.<br>

[푸리에 스펙트럼 (=magnitude, 매그니튜드)]

![푸리에 스펙트럼과 좌표계](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/%ED%91%B8%EB%A6%AC%EC%97%90%20%EC%8A%A4%ED%8E%99%ED%8A%B8%EB%9F%BC%EA%B3%BC%20%EC%A2%8C%ED%91%9C%EA%B3%84.JPG?raw=True)

-푸리에 스펙트럼은 저주파 영역은 매우 큰 값을 갖고 고주파 영역은 0에 가까운 값을 갖는다.<br>
1)푸리에 스펙트럼을 그대로 이미지로 시각화하면 검은 바탕 위에 흰점하나만 존재하여 log를 취하면 (b)처럼 모서리로 갈수록 값이 높아지는 스펙트럼 형태여서, (c)처럼 원점 중심에 오도록 shift형태의 이미지를 사용하는 것이다.

[푸리에 페이즈 (=phase)]

-페이즈는 파형 주기의 시작점의 각도 또는 어느 한 순간의 위치이다.
-원본 신호를 FT로 분해하면 각 주기성분의 시점이 어딘지(각도)를 나타낸다.

하기의 그림은 동일한 파란색 주기성분 신호의 위치들이 변하면 빨간색 신호 생성이 다름을 보여준다.
![푸리에 스펙트럼과 좌표계](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/fourier%20phase.png?raw=True)

아래 그림에서 (a)는 원본 이미지, (b)는 푸리에 스펙트럼을 보존하고 페이즈(phase)를 랜덤(random)하게 했을 때의 역변환 결과, (c)는 페이즈(phase)를 보존하고 스펙트럼을 랜덤하게 했을 때의 역변환 결과이다. 결과를 보면 이미지의 푸리에 변환에서 스펙트럼(spectrum)보다 페이즈(phase)에 보다 더 중요한 정보가 포함되어 있음을 확인할 수 있다.
![푸리에 스펙트럼과 좌표계](https://github.com/tenjumh/GraduateSchool/blob/master/Study/Fourier%20Transform/dscription%20image/fourier%20transform_magnitude_phase.png?raw=True)

```python
import matplotlib.pyplot as plt
from skimage import io, filters, feature
import numpy
from scipy import fftpack, signal
import scipy

def gkern(kernlen, nsig):
    # Return 2D Gaussian Kernel
    gkern1d = signal.gaussian(kernlen, std=nsig).reshape(kernlen, 1)
    kernel = numpy.outer(gkern1d, gkern1d)

    return kernel

# Load image file
fpath = "C:/Users/fxk/PycharmProjects/tenjumh/Computer Vision/191018/"#'./PycharmProjects/tenjumh/Computer Vision/191018/'
image = io.imread(fpath + 'character.tif').astype('float64')
# Fourier Transform
imageFFT = fftpack.fft2(image)
imageFFT_abs_log = scipy.log10(1+scipy.absolute(imageFFT))
imageFFT = fftpack.fftshift(imageFFT)   
imageFFT = scipy.log10(1+scipy.absolute(imageFFT))

plt.figure(figsize=(10, 10), dpi=150)
plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('1)OriginalImage', {'fontsize': 10})
plt.subplot(3, 3, 2)
plt.imshow(imageFFT_abs_log, cmap='gray')
plt.title('2)Fourier', {'fontsize': 10})
plt.subplot(3, 3, 3)
plt.imshow(imageFFT_log, cmap='gray')
plt.title('3)Fourier Shift', {'fontsize': 10})

plt.show()
```
