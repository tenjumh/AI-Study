## 1. plt.subplot()
- subplot(m, n, i) : m행, n열, i번째
<br>

  subplot(2,2,1)   |   subplot(2,2,2)<br>
  :----------------------------------------------<br>
  subplot(2,2,3)   |   subplot(2,2,4)<br>  

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,50)
y1 = np.cos(4*np.pi*x)
y2 = np.cos(4*np.pi*x)*np.exp(-2*x)
```

```python
plt.subplot(1,1,1)
plt.plot(x, y1, 'r-*', lw=1)
plt.ylabel('$sin(\pi x)$')
```

```python
plt.subplot(1,1,1)
plt.plot(x, y1, 'r-*', lw=1)
plt.ylabel('$sin(\pi x)$')
plt.axis([0, 1, -1.5, 1.5])    #plt.axis([x축 start, x축 end, y축 start, y축 end])
```

```python
plt.subplot(2,1,1)
plt.plot(x, y1, 'r-*', lw=1)
plt.ylabel('$sin(\pi x)$')
plt.axis([0, 1, -1.5, 1.5])    #plt.axis([x축 start, x축 end, y축 start, y축 end])

plt.subplot(2,1,2)
plt.plot(x, y2, 'b--o', lw=1)
plt.ylabel('$e^{-2x}sin(4\pi x)$')
plt.axis([0, 1, -1.5, 1.5])

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x, y1, 'r-*', lw=1)
plt.ylabel('$sin(\pi x)$')
plt.axis([0, 1, -1.5, 1.5])    #plt.axis([x축 start, x축 end, y축 start, y축 end])

plt.subplot(2,2,4)
plt.plot(x, y2, 'b--o', lw=1)
plt.ylabel('$e^{-2x}sin(4\pi x)$')
plt.axis([0, 1, -1.5, 1.5])

plt.show()
```
